# 4. Statistical Analysis:

### Stating Ho, Ha
### Testing Hypothesis

```python
#Null hypothesis (Ho) = No difference between Means
#Alt hypothesis (Ha) = Difference between the two means
```

```python
from statsmodels.stats.weightstats import ztest

s1= df1.loc[df1.OperatingState == 'Operating', 'Funds_in_million']
s2= df1.loc[df1.OperatingState == 'Closed', 'Funds_in_million']

alpha = .05

t_stat, p_val = ztest(s1,s2)

if p_val <= alpha:
    print ('We Reject the Null Hypothesis as ''Value of P is %1.4f' %p_val, 'which is lower than alpha')
else:
    print ('We Accept the Null Hypothesis as ''Value of P is %1.4f ' %p_val, 'which is greater than alpha')
```

### Percentage of winners that are still operating and percentage of contestants that are still operating


```python
winners = df2.Result.value_counts()['Winner']
contestants = df2.Result.value_counts()['Contestant']

contestants_operating = df2.OperatingState[df2.Result == 'Contestant'].value_counts().loc['Operating']
winners_operating = df2.OperatingState[df2.Result == 'Winner'].value_counts().loc['Operating']

percent_winner_operating = (winners_operating/winners)*100
percent_contestant_operating = (contestants_operating/contestants)*100

print ('Number of Winners is', winners)
print ('Number of contestants is', contestants)
print ('Number of Winners that are operating is', winners_operating)
print ('Number of contestants that are operating is', contestants_operating)
print ('Percentage of Winners that are still operating is %1.2f' %percent_winner_operating,'%')
print ('Percentage of Contestants that are still operating is %1.2f' %percent_contestant_operating,'%')

```

### Hypothesis comparing the proportion of companies that are operating between winners and contestants


```python
#Null hypothesis, Ho = proportion of companies that are operating between winners and contestants is same
#Alternate hypothesis, Ha = proportion of companies that are operating between winners and contestants is different
```

```python
from statsmodels.stats.proportion import proportions_ztest

tstat, p_val = proportions_ztest([contestants_operating, winners_operating], [contestants, winners])
alpha = 0.05

if p_val <= alpha:
    print ('We Reject the Null Hypothesis as ''Value of P is %1.4f' %p_val, 'which is lower than alpha')
else:
    print ('We Accept the Null Hypothesis as ''Value of P is %1.4f ' %p_val, 'which is greater than alpha')
```

## CONCLUSION: There is not much difference between Winners and Contestants in being Operational
